### TimeResampling

**'A' (Year End Frequency):  
    1. Resamples the data to the end of the calendar year.  
    2. Ignores weekends and holidays.  
    3. Each year is treated the same, regardless of whether it's a business or calendar year.**

```python
df.resample(rule='A').max()    ## Year end frequency
```

```python
df.resample(rule='A').max()['High'].plot() 
```

```python
df.loc['2015-01-01':'2015-12-31'].max()
```

```python
df.resample(rule='A').min()
```

```python
df.loc['2015-01-01':'2015-12-31'].min()
```

**'BA' (Business Year End Frequency):**  
**1. Resamples the data to the end of the business year, where a business year typically ends on the last weekday of December.**  
**2. Takes into account only business days (weekdays) and skips weekends and holidays.**    
**3. Useful in financial contexts where weekends and holidays are not considered as trading days.**  

```python
df.resample(rule='BA').max()
```

**'QS' (Quarter Start Frequency):    
1.Resamples the data to the start of each calendar quarter.  
2.Considers all days, including weekends and holidays.  
3.Each quarter is determined by the calendar, regardless of business days.**

```python
df.resample(rule='QS').mean()['High'].plot()
```

```python
df.loc['2015-01-01':'2015-03-31'].mean()
```

**'BQS' (Business Quarter Start Frequency):  
1.Resamples the data to the start of each business quarter.  
2.Takes into account only business days (weekdays) and skips weekends and holidays.  
3.Useful in financial contexts where weekends and holidays may not be considered as part of the business quarter.**

```python
df.resample(rule='BQS').mean()['Open'].plot()
```
